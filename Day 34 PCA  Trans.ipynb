```python
# Load Data Set

df=pd.read_csv('trans_us.csv',index_col=0,thousands=',')
display(df)
```

```python
# Update Row and Column Headings

df.index.names=['Country']
df.columns.names=['Years']
display(df)
```

```python
# Check for Null Values

display(df.isna().sum())
```

```python
# PCA with n-components = None

from sklearn.decomposition import PCA
pca=PCA()
pca.fit(df)
print(pca)
```

```python
# Transform and display data

pajju=pca.transform(df)
display(pd.DataFrame(pajju))
```

```python
# Display the variance Percentage

print(pd.DataFrame(pca.explained_variance_ratio_))
```

```python
# Create Model

from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(df)
print(pca)
```

```python
# Create PCA with two variables

pajju=pca.transform(df)
display(pd.DataFrame(pajju))
```

```python
# Convert output to Data Frame

pajju2=pd.DataFrame(pajju)
pajju2.index=df.index
pajju2.columns=['PC1','PC2']
display(pajju2.head(50))
```
